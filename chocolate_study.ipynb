### Visualization 

```python
import matplotlib.pyplot as plt 
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, drop='first')

X = dataset_dropped_na['Rating']
print(dataset_dropped_na['Rating'].info())

# Histogram of rating
plt.hist(X)
plt.show()
```

#### observations 

The histogram shows a unimodal distribution centered around a rating value of roughly 3. The tallest bar rises slightly above 600, meaning most observations fall close to this central value. The spread of the data is fairly narrow: nearly all ratings lie between 2.5 and 3.5, with very few values outside that range. The shape leans slightly to the right, suggesting a mild right‑skew, but the overall pattern is still tightly clustered.


```python
# Converting float to int 
dataset_dropped_na.info()
convert_column= []
for i in dataset_dropped_na['Rating']:
    convert_column.append(int(i))
print(convert_column)

# Histogram of the new dataset
plt.hist(convert_column)
```

```python
convert_column1 = []

# Converting a percentage in str to int 
for i in dataset_dropped_na['Cocoa\nPercent']:
    convert_column1.append(int(i[:2]))
print(convert_column1)

# Scatter diagram
plt.scatter( dataset_dropped_na['Rating'],convert_column1, alpha= 0.1)
```

Most points cluster tightly between ratings of 2.5 and 3.5, where the percentages predominantly fall within the 60–80% range, creating a darker, denser region due to overlapping semi‑transparent points. Outside this central band, the data becomes more dispersed, with fewer observations at both lower and higher percentages. Overall, the plot suggests that while percentage values vary widely for each rating, the dataset is heavily concentrated around mid‑level ratings, indicating a weak or non‑linear relationship between the two variables.

#### We will normalize using the Min = Max Normalizetion (0 - 1 scale)



```python
# Normalization 

ratings = dataset_dropped_na['Rating']

min_val = dataset_dropped_na['Rating'].min()
max_val = dataset_dropped_na['Rating'].max()

# Formular
normalized = (dataset_dropped_na['Rating'] - min_val) / (max_val - min_val)

print(normalized)

# Histogram plot 
plt.hist(normalized)
```

```python
# List of companies ordered by their average score (averaged over each company’s reviews).
dataset_dropped_na.groupby(dataset_dropped_na.columns[0])['REF'].mean().sort_values(ascending = False)
```

```python
# Encoding the companies column using the OneHot encoded 
companies_name_encoded = ohe.fit_transform(dataset_dropped_na.iloc[:, :1])

#Creating a column of headers 
companies_encoder_column = [f'company_{i}' for i in ohe.categories_[0]]

#Building a dataset
companies_encoded_dataset = pd.DataFrame(companies_name_encoded, columns= companies_encoder_column[1:])

#Here we concatenate with the privious clean dataset in file
df = pd.concat([dataset_dropped_na , companies_encoded_dataset], axis = 1)
print(df)
```

We have a bigger dataset with 1795 rows and 424 columns

```python
# Encoding the location column using the OneHot encoded 
location_encoded = ohe.fit_transform(dataset_dropped_na.iloc[:, -4:-3])

#Creating a column of headers as follow location_(actual location) 
location_encoded_columns = [f'location_{i}' for i in ohe.categories_[0]]

#Building a dataset with location_encoded_columns as headers and location_encoded as data
location_encoded_dataset = pd.DataFrame(location_encoded, columns= location_encoded_columns[1:])

#Here we concatenate with the privious clean dataset in file
df = pd.concat([df, location_encoded_dataset], axis = 1)
print(df)
```

Our dataset is now 1795 rows and 542 columns, with location and companies columns OneHot Encoded.
